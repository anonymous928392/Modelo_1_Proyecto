```python
import pandas as pd

df = pd.read_csv('train.csv')
df.dropna()
df.head()
```

#Cantidad de valores nulos por columna

```python
df.isnull().sum()
```

# Estudiantes por programa academico de mayor a menor

```python
students_per_class = df['ESTU_PRGM_ACADEMICO'].groupby(df['ESTU_PRGM_ACADEMICO']).count().sort_values(ascending=False)
students_per_class
```

 # Carrera más común por Departamento

```python
students_per_department = df.groupby(['ESTU_PRGM_DEPARTAMENTO', 'ESTU_PRGM_ACADEMICO']).size().reset_index(name='Cantidad')
sort_students_per_department = students_per_department.sort_values(['ESTU_PRGM_DEPARTAMENTO', 'Cantidad'], ascending=[True, False])
most_common_program = sort_students_per_department.groupby('ESTU_PRGM_DEPARTAMENTO').first().reset_index()
most_common_program
```

# Número de estudiantes por departamento cuyo rendimiento global fue alto

```python
rendimiento_department = (df[df['RENDIMIENTO_GLOBAL'] == 'alto']
                          .groupby(['ESTU_PRGM_DEPARTAMENTO', 'RENDIMIENTO_GLOBAL'])
                          .size()
                          .reset_index(name='Cantidad')
                          .drop(columns='RENDIMIENTO_GLOBAL')
                          .sort_values('Cantidad', ascending=False))

rendimiento_department
```
